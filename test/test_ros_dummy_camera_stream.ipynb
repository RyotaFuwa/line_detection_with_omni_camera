```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
element_spec = {
    "image": tf.TensorSpec(shape=(None, 128, 128, 3), dtype=tf.float32, name=None),
    "label": tf.TensorSpec(shape=(None, 128, 128, 1), dtype=tf.float32, name=None),
    }
ds = tf.data.experimental.load('data/datasets/train_ds', element_spec)
```

```python
images = list(ds.as_numpy_iterator())
```

```python
image_list = []
for d in images:
    image_list.append((d['image'][0] * 255).astype('uint8'))
```

```python
model = tf.keras.models.load_model('data/models/2020-10-17.h5')
```

```python
pred = model.predict(np.expand_dims(image_list[1], 0) / 255.0)
```
